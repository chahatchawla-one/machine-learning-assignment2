# Load linearly separable

```python
dataset = pd.read_csv("data/Data.csv")
dataset_x = dataset.drop("label", axis = 1).values
dataset_y = dataset["label"].values

x_train = dataset_x[:9000]
y_train = dataset_y[:9000]
x_test = dataset_x[9000:]
y_test = dataset_y[9000:]

plt.scatter(x_train[:1000, 0], x_train[:1000, 1], c = y_train[:1000], marker = ".")
```

# Perceptron

```python
classifier = Perceptron()
predictions = classifier.Fit(x_train, y_train, epochs = 5)
print("Accuracy =", Accuracy(predictions, y_train))
```

```python
fig = plt.figure()
plt.scatter(x_train[:1000, 0], x_train[:1000, 1], c = y_train[:1000], marker = ".")
w_ = classifier.weights.reshape(3)
plt.plot([0, 1], [-w_[0] / w_[2], -(w_[0] + w_ [1]) / w_[2]])
```

```python
from sklearn.svm import SVC

classifier = SVC(kernel = "linear")
classifier.fit(x_train, y_train)
predictions = classifier.predict(x_test)
Accuracy(predictions, y_test)
```
